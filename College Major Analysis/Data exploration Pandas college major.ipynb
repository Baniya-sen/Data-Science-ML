```python
# 1.
highest = clean_df['Mid-Career Median Salary'].idxmax()
print(clean_df['Undergraduate Major'][highest])
clean_df['Mid-Career Median Salary'][highest]
```

```python
# 2.
lowest = clean_df['Starting Median Salary'].idxmin()
print(clean_df['Undergraduate Major'][lowest])
clean_df['Starting Median Salary'][lowest]
```

```python
# 3.
lowest_m = clean_df['Mid-Career Median Salary'].idxmin()
print(clean_df['Undergraduate Major'][lowest_m])
clean_df['Mid-Career Median Salary'][lowest_m]
```

Subtract whole column

```python
risk_factor = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
# Or use- clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
# del clean_df['Low Risk']
clean_df.insert(5, 'Risk Factor', risk_factor)
clean_df.head()
```

Sort whole column and print only 2 columns

```python
sorted_df = clean_df.sort_values('Risk Factor')
sorted_df[['Undergraduate Major', 'Risk Factor']].head()
```
